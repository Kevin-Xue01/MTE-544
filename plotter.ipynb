```python
import pandas as pd
from matplotlib import pyplot as plt

ground_truth_file = 'csv/0/CIRCLE/EKF_odom.csv' # x,y,v,w # This is ground truth
odom_file = 'csv/0/CIRCLE/EKF_noisy_odom.csv' # x,y,v,w
ekf_file = 'csv/0/CIRCLE/EKF_robotPose.csv' # x,y,th,stamp
imu_file = 'csv/0/CIRCLE/imu.csv' # x,y,th,stamp
ukf_file = 'csv/0/CIRCLE/UKF_robotPose.csv'

# Read EKF estimate CSV file
ekf_robotPose = pd.read_csv(ekf_file)
# ukf_robotPose = pd.read_csv(ukf_file)
ground_truth = pd.read_csv(ground_truth_file)
odom = pd.read_csv(odom_file)

# Plot odometry and EKF estimate trajectories
plt.figure(figsize=(10, 6))
plt.plot(ground_truth['x'], ground_truth['y'], label='Ground Truth', color='red', alpha=0.7)
plt.plot(odom['x'], odom['y'], label='Odom', color='blue', alpha=0.7)
plt.plot(ekf_robotPose['x'], ekf_robotPose['y'], label='EKF Estimate', color='green', alpha=0.7)
# plt.plot(ukf_robotPose['x'], ukf_robotPose['y'], label='UKF Estimate', color='green', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Robot Trajectory: Ground Truth vs UKF Estimate')
plt.legend()
plt.grid()
plt.show()
```
